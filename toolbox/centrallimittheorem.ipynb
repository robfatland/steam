```python
from matplotlib import pyplot as plt
import numpy as np

def gauss(x, mu, sigma):
    g = 1./(sigma*np.sqrt(2.*np.pi))
    arg = (-1/2)*((x-mu)/sigma)**2
    return g*np.exp(arg)

xx = []      # ordinates
g  = []      # abscissa
m  = -5      # mean offset
s  = 3.2     # standard deviation sigma

x0, x1, dx = -20, 20, .01      # x-axis range and increment

for x in np.arange(x0, x1, dx):
    g.append(gauss(x, m, s))
    xx.append(x)
plt.plot(xx, g)
```

```python
from random import randint

def coin(): return float(randint(0,1))

def sample_mean(n):
    sample_sum = 0.
    for i in range(n): sample_sum += coin()
    return sample_sum/n

h = np.zeros(101)                        # histogram
n_samples = 10000
tosses = 30
for i in range(n_samples):
    h[int(sample_mean(tosses)*100)] += 1
    
plt.plot(h)
```
