### Electronegativity


Context, paraphrasing from Wikipedia: Electronegativity is a *tendency* to attract or accept shared electrons 
when forming chemical bonds. Electronegativity is affected by both atomic number (in terms
of which column of the periodic table, i.e. valence electron count) and the radial distance
of the atomic (or molecular?) valence electrons. Higher electronegativity: stronger electron
attraction. Electronegativity relates to bond energy. A chemical bond has a strength -- the
energy needed to break the bond -- and this strength is related to the shared versus separated
nature of the distribution of electrons. At the shared end of the spectrum we have covalent bonds
and when the electrons are separated we have ionic bonds. The term electropositivity is an
opposite of electronegativity: it characterizes an element's tendency to donate valence electrons.


Here is the functional result of all this electronegativity context: First, the right-hand column
of the periodic table is noble gasses that have all their outer shells filled so they are not in
the bonding business. Ignoring the right-most column: Electronegativity increases from left to right
in the periodic table in the column sense; and electronegativity increases from higher to lower 
energy levels, i.e. in the periodic table row sense. As a consequence, the most electronegative 
atom is ...drumroll please... Fluorine. This acts as the origin for electronegativity. 
If we are two chose the *least* electronegative atom: It will be the one *furthest* from Fluorine on 
the periodic table.


Example: The molecule $NO_2$ features oxygen and nitrogen. Nitrogen is further from Fluorine so
it is less electronegative than oxygen. 


### The Five Rules of Lewis structures


These rules are the *basic set*. There are advanced rules; but I will kick that can down the road for now.


0. Draw the individual atoms with their dots (this is optional... 'know your actors')
    - ...then follow the rules below on a clean slate
1. Find the total number of valence electrons for a given molecule.
    - For water we should get 8 valence electrons: 6 from oxygen and one each from the hydrogens
    - For a negative ion we have an atom or molecule carrying additional charge
        - Example: $CN^-$: 4 valence electrons for Carbon, 5 for Nitrogen, 1 for the ionic charge: 10 total
2. Establish the "center" atom of the Lewis structure
    - For 2-atom molecules there is no center
    - For three or more atoms there is a center
        - Place the least electronegative atom at the center
        - Place the other atoms on the outside
        - Caveat rule: Hydrogen is a special animal
            - Hydrogen is always placed on the outside of a Lewis structure
        - Example: $P\;Cl_3$ will have a phosphorous atom at the center (check this)
3. From this arrangement of atoms: Put 2 electrons between atoms to make covalent $\sigma$ bonds
    - Simple example: $H\;Cl$ has 8 valence electrons (Hydrogen: 1, Chlorine: 7)
        - It has two atoms so there is no center
        - Place the first two valence electrons like **H:Cl** to make the $\sigma$ bond
            - Not now but at the conclusion: Re-draw each two dot bond ':' as a dash '-'
    - We are working from a valence electron budget of 8 so 6 remain
4. Complete the "8 electron" octets: First for the outside atoms, then for the center atom 
    - Hydrogen doesn't have an octet. Its outer shell can carry 2 electrons.
        - In the above **H:Cl** example Hydrogen has 2 from the covalent bond: All set
        - The Chlorine gets the remaining 6 electrons as 3 lone pairs: 8 total: All set
        - We are done with the HCl structure

```
    ..
H - Cl :
    ..
```
        

5. If the central atom does not have an octet: Move electrons from outer atoms to form double or triple bonds
    - This happens when we run out of valence electrons without filling the octets
        - There could be the alternative issue of too *many* remaining electrons
    - The example directly following shows Rule 5 in action


### Lewis structure building in practice


#### Simple example: $O_2$


- Rule 1: This is two Oxygen atoms, same electronegativity, 12 total valence electrons
- Rule 2: Two atoms so no 'center atom' as such
- Rule 3: One $\sigma$ bond (2 electrons: 10 remaining)
- Rule 4: Complete the octet for one Oxygen (6 electrons: 4 remaining)
    - Try to complete the octet for the second Oxygen (4 electrons: 0 remaining)
    - Now the second oxygen is 6 out of 8 octet electrons and rule 5 applies
- Rule 5: Move two electrons from the first oxygen to create another valence bond
    - This would be a first $\pi$ bond
    - Using dots we have `O::O`. Using lines we have `O=O`.  
- Now the octets are full for both oxygen atoms and the Lewis structure is complete
    - It is best to include all the valence electrons in the diagram, as in
    
    
```
  ..     ..
: O   =  O :
```

#### A more complicated example: Sulfur dioxide, $SO_2$.

- Rule 1: 18 valence electrons
- Rule 2: Sulfur is farther from Fluorine than Oxygen so it goes in the center
- Rule 3: The covalent bonds use up 4 electrons, produce O:S:O, 14 valence electrons remain
- Rule 4: 6 electrons to each Oxygen to complete outer octets and 2 electrons remain
    - The sulfur atom has 4 electrons in the two $\sigma$ bonds
    - Add the 2 remaining valence electrons to sulfur for a total of 6
- Rule 5: An asymmetric steal: Choose the left oxygen and move two electrons to the sulfur bond
        - This is now a double ($\pi$) bond and we have O::S:O
        - The sulfur now has a completed octet, as do the two oxygens
        - ...so we might imagine we are done...

### Advanced material


There are two topics that we have not incorporated yet: Formal charge and resonance structures. 
(I did mention the latter briefly for the Ozone molecule.)


*Formal charge* in covalent bond theory is a charge assigned to an atom within a molecule.
It is defined as a difference between the Lewis structure state of the atom and the state
of the same atom 'by itself' (or 'free from the molecule (or 'free')).


$q* = V - L - B/2$


- $V$ is the number of valence electrons for the atom
- $L$ is the number of non-bonding electrons assigned to the atom in the Lewis structure
- $B$ is the total number of electrons shared in bonds with other atoms in the molecule


$B$ is divided by two as the 2-electron covalent bond is considered *shared*.


*Resonance structures* are related to a dichotomy: On the one hand the mechanical rules 
for a Lewis structure tend to imply a *location* for an electron; and on the other hand
the rules of quantum mechanics allow for electrons to have a distributed existence that
defies the classical concept of an object having a location. 


Here is my attempt at an unhelpful (helpful?) conversation on this topic:


A: I have solved the Lewis structure for Ozone; but I have a problem.


N: What is your solution? 


A: It is `O::O:O` which you will notice features both a double and a single covalence bond.


N: And what is your problem?


A: I could just as easily say that the solution is `O:O::O`... so how does the molecule decide?


N: Ah, well as a practitioner of quantum mechanics I can assure you that the molecule is under no obligation to chose.


A: But it seems to me that the two solutions are equally valid; so perhaps the 'choice' simply happens at random.


N: Not at all. Rather the ozone molecule exists in a dual state: A superposition of your two solutions. That is: The molecule is not in strictly state `O::O:O` or state `O:O::O`. It exists in both at once.


A: Doesn't that make it heavier? Or set up a gateway to the Multiverse? Or something? 


N: It is a bit of a cop out, but I will just claim that the mathematical solution for the configuration of electrons in this molecule is a sort of hybridization of these two symmetric configurations; hence I say "both".



## Valence Shell Electron Pair Repulsion (VSEPR) theory

The nice thing about VSEPR is that it starts from Lewis structures. In particular remember
that we have lone pairs in our Lewis structures in addition to 2-electron valence bonds. 
The goal is to use some heuristic (rule of thumb) arguments to come up with molecular
structure. This is sometimes done using a sub-method referred to as the AXE method. So
my goal here is to describe the VSEPR rationale in an arm-waving sort of way, to describe
a key number we calculate called the *steric* number, and then to proceed to giving the 
AXE method to arrive at some example molecular structures. 


One idea to read and ponder before jumping in here: Lewis structure theory is all about
building molecules from an economy of electrons: *Where do the electrons go?* In contrast
VSEPR is ultimately about the geometry of molecules containing three or more atoms, a
geometry of a 3D tree graph where the **edges** are covalent bonds. 


### Part 1 A conception of orbitals

In VSEPR theory we have a metaphor of orbitals as ballons: Volumes that are blob-like,
that keep with the notion that electrons exist in a distributed (pdf-like) space. We can
say that *orbitals* are regions of space where one has a high probability of "finding" an
electron and VSEPR is concerned with the geometric configuration of these blobby balloons,
these orbitals. The balloons are tied together at the atom's nucleus; each can contain two
electrons (spin up and spin down, whatever that means) and because electrons are all 
negatively charged these orbitals repel one another. Now let's go a little further with
a taxonomy of orbitals. 


Each lone pair orbital we presume contains two electrons; and it exists relatively close
to the nucleus because there is no other atom drawing it out *away* from the nucleus. 
Because this lone pair orbital is compact it has a highly repulsive nature. But we try 
and be encouraging and kind despite this observation.

Moving on to $\sigma$-bond orbitals: These are also occupied by two electrons. They are
the longest type of covalent bond, the opposite of lone pair orbitals being compact. We
can say that they are the weakest of the three valence bonds we learned about in Lewis 
structures. 


So far so good; now for the twist. 


In VSEPR theory: Double $\pi$-bonds are counted as a single chemical bond with capacity 
for four electrons. Notice this is not the same as a four-electron *orbital* because I
am not sure that would be permitted language. The reason is that orbitals are related
more closely to quantum mechanics and quantum mechanics is very strict about electron
double occupancy. So instead of using the word *orbital* I will stick with a (double)
valence bond which is stronger (and therefore shorter -- rubber band analogy) than
a single bond. Respectively $\pi$ and $\sigma$ bonds. 


Finally for VSEPR we also have triple $\pi$-bonds which are stronger/shorter still. 
But these are not as compact as lone pair orbitals. So the respective repulsion
of other bonds/orbitals is maximum for lone pairs, then triple-$\pi$, then double-$\pi$,
then single $\sigma$ valence bonds. This comes into play as a secondary factor on
the molecular geometry we are attempting to conjure up. 


If all of the above business about **E**lectron **P**air **R**epulsion was *secondary*:
What is the *primary* thing? Answer: The *steric number* which we now propose to 
define. 

### Part 2 Steric number


The steric number is calculated for a central atom in a molecule. The leaf atoms are
important in this counting process but they do not have their own steric numbers.


Counting the steric number: Single $\sigma$ valence bonds count 1. A double $\sigma, \pi$ bond counts as 1. 
A triple $\sigma, \pi, \pi$ bond counts as 1. A lone pair counts as $1. Considering a central (not leaf)
atom: The sum steric number dictactes the bond geometry where each bond/lone pair repels the others.
Hence a carbon atom surrounded by four valence bonds to Hydrogen will have a tetrahedral shape. 

## Linear Combinations of Atomic Orbitals (LCAO)


This topic originates from the notion that orbitals describe the space occupied (in a cloudy or probabilistic sense)
by an electron; and in fact any such orbital can carry 1 or 2 electrons (of opposite spin) and no more.
This can be stated another way: We describe an electron in an atom or molecule in terms of three-dimensional
volumes called orbitals. We begin with atoms where orbitals are centered on the nucleus; but we find when atoms
bind together that orbitals can be distributed across two or even more nuclear 'origin points'. 


These notes are concerned with the atomic orbitals specifically of Hydrogen. Because hydrogen is
a simple enough system its orbitals can be solved for mathematically. Chemists then go on to say
'suppose all atoms in a given molecule have at their disposal Hydrogen orbitals centered on their
respective nuclei... albeit maybe scaled because of the larger mass / more protons in the nucleus /
presence of inner shell electrons. But the basic concept is the Hydrogen orbitals act like basis
vectors in linear algebra, like i-hat, j-hat, k-hat: They can be combined using coefficients into
a linear combination. Very well: Let's take a further leap supposing that *molecular* orbitals are
formed as linear combinations of atomic orbitals: Again centered on the multiple atomic nuclei. 
This is the notion of LCAO.


In terms of first-year chemistry: Since we see the phrase "linear combinations" we imagine that the
rules of linear algebra might apply. This is 100% correct: The primary method for solving an LCAO 
problem is called the **Hartree-Fock method** and it involves diagonalizing a matrix of coefficients
in a repetition cycle that will eventually minimize the energy of the molecule. The rationale is 
that in nature things tend to head towards a minimal energy state. In atomic and molecular systems
this is called the ground state.

### Anti-bonding orbitals


[Wikipedia overview](https://en.wikipedia.org/wiki/Antibonding_molecular_orbital)


* Anti-bonding weakens the chemical bond between two atoms.
* It is not a sure thing to break the molecule apart; but it does affect the total energy.
* Anti-bonding (high energy) orbitals fill only after bonding (lower energy) orbitals fill


Consider two hydrogen atoms far apart: They have one electron each, occupying respective 1s orbitals. 
Now they move close together and the two atomic orbitals transform into two molecular orbitals: 1s and 1s*.
These two new orbitals conserve the total orbital count. 


The two electrons can both drop down into the 1s molecular orbital. This lowers the total energy
in comparison to that of the two free atoms; so $H_2$ is a stable bound state of molecular hydrogen.
The 1s* orbital contains no electrons. 


Now do the same thing with Helium: Both 1s and 1s* are occupied (4 electrons total). The anti-bonding
1s* has a big energy gap relative to the energy of two separate Helium atoms. The bonding 1s orbital
has a less-big negative energy gap relative to the energy of two separate Helium atoms. In sum the 
$He_2$ molecule has more energy than He + He: The Helium molecule is unstable and breaks apart.



### $\sigma$ and $\pi$ bonds


Note: nodal planes are planes that cut orbitals at the point where they become zero. A p orbital
has a nodal plane passing through the nucleus, for example.


From wikipedia: 

#### $\sigma$ bonds


* $\sigma$ bonds are the strongest type of covalent bond
    * This is due to a maximization of overlap between bonding orbitals along the bond axis
    * This is in contrast to $\pi$ bonds that overlap (less) in a side-by-side manner
* $\sigma$ bonds involve head-on matching of orbitals from two atoms along the inter-nuclear (bond) axis
    * This axis is often labeled the z-axis
* $\sigma$ bonds are symmetric (unchanged by / tolerant of) rotation about the bond axis
    * This is, again, in contrast to $\pi$ bonds
* $\sigma$ bonds are commonly of type **s+s**, **pz+pz**, **s+pz** and (invoking d-orbitals) **dz2+dz2**
    * These are hybridized orbitals
* $\sigma$ bond hybridized molecular orbitals do not have a nodal plane (where they go to zero)
* $\sigma^*$ anti-bonding molecular orbitals *do* have a nodal plane
    * This resides between the two nuclei and is perpendicular to the z-axis


#### $\pi$ bonds

* $\pi$ bonds are built from overlap of parallel p orbitals: Side by side dumbells
    * Another good comparison is two boats rafted together
    * Wikipedia has a helpful remark here:
        * "Molecular fragments joined by a pi bond cannot rotate about that bond without breaking the $\pi$
        * bond because rotation involves destroying the parallel orientation of the constituent p orbitals."
* $\pi^*$ anti-bonding orbitals have two nodal planes
    * one is a nodal plane through both nuclei perpendicular to the atomic p orbitals
    * one is in the xy-plane between the two nuclei

 
#### Delocalization and resonance


Delocalization refers to electrons that are not associated with strictly one or two atoms; 
but may have an orbital that extends across three or more atoms. This is related to the concept
of resonance and is seen in benzene (6-Carbon) rings, schematically as a circle inside the hexagon.


Note: An isomer is an alternative structure for a molecule using the same atoms. This is *not* the
same thing as resonance structures. Resonance structures vary from one another in placement of 
electrons in the bond assembly *without* changing the arrangement of the atoms i.e. the graph of 
the atoms within the molecule.

I interrupt these notes to tell you a concept that brings together some
things you already know. This concerns the statement "The square of the
wave function is the probability density function (pdf)." This is not
going to be rigorous but it may help a bit. The pdf defines "where the
electron is likely to be" as a volume of space associated with the 
molecule as "the origin". But how can two atomic orbitals be added
together destructively to create nodal points where the wave function
goes to zero? The answer is that the wave function is ***complex valued***.
Let's suppose that at a particular point the "a" atomic orbital or
wave function has the value $0.3 e^{i\frac{\pi}{6}}$ and the "b" atomic
orbital or wave function has the value $0.3 e^{i\frac{-5\pi}{6}}$. 
If we square these and take the absolutely they both give us 0.09
which is a nice positive number. But if we superimpose the two 
wave functions first by adding the two values: They happen to be
"out of phase" so their sum is zero and the resulting square has
magnitude zero. So wave functions are complex valued and that is how
the math gets us into *constructive* and *destructive* forms of 
interference.


It is a philosophical matter to ask whether the atoms decided to play
dice using complex numbers or if the atoms are doing something else and
this complex number business is simply a convenient method invented
by the humans to describe that behavior.

## Types of Reactions


For our purposes (avoiding nuclear chemistry) a chemical reaction will involve a redistribution of 
electrons (and implicitly chemical bonds) together with a change in *energy*. The chemical reaction
starts with some *reactants* (aka *reagents*) and arrives at some *products*. 


Reactions occur at some rate which is governed by quantity of reactants and temperature... and 
by other things, particularly presence of a *catalyst*. 


Reactions that produce heat are *exothermic*. 


Reactions that require heat are *endothermic*. 


**I interrupt to say: More properly *energy*, usually in the form of *heat* ($\Delta$) 
or *light* ($h\nu$).**


Reactions in which oxidation and reduction occur are called *redox* reactions. Antonym: *non-redox*.


Most simple redox reactions are in one of three sub-categories: *Combination*, *Decomposition*, or *Single Replacement*.


Chemical reactions may involve any combination of four basic characteristics:


- Evolution of a gas
- Precipitation
- Change in temperature
- Change in state, for example melting


Equilibrium reactions are chemical reactions that can proceed in both directions:
Reactants to products and vice versa. 


A reaction equation features stoichiometric balance: The quantities of each atom seen
on opposite sides of the reaction arrow are equal. This can be achieved by solving for integer 
coefficients in the manner of a Diophantine problem.
